```python
import matplotlib.pyplot as plt
import numpy as np

NUM_DATAPOINTS = 100000
MIN = -5
MAX = 5
SIGMA = 1

def second_derivative(xs, sigma):
  a = (xs**2 - sigma**2)/(np.sqrt(2*np.pi)*(sigma**5))
  return a * (np.e ** (-(xs**2)/(2*(sigma**2))))

xs = np.linspace(MIN, MAX, num=NUM_DATAPOINTS)

ys = second_derivative(xs, SIGMA)

fig = plt.figure()
ax1 = fig.add_subplot()
ax1.set_ylabel(r'$\frac{d^2g_{\sigma}(x)(x)}{dx^2}$')
ax1.set_xlabel('x')
ax1.set_title('2nd Derivative of the Normal Gaussian')
ax1.plot(xs, ys)
plt.show()
```

```python
k = [1.2, 1.4, 1.6, 1.8, 2.0]

def gaussian(xs, sigma):
    return (1/(np.sqrt(2*np.pi)*sigma)) * np.e ** (-(xs ** 2)/(2 * (sigma ** 2)))
gauss_1 = gaussian(xs, 1)

plt.figure(figsize=(16, 8), dpi=80)
for i in range(len(k)):
    D = (gaussian(xs, k[i]) - gauss_1) / (k[i] - 1)
    
    plt.subplot(2, 3, i+1)
    plt.plot(xs, ys)
    plt.plot(xs, D)
    if i == 3 or i ==4:
        plt.xlabel('x')
    if i == 0 or i ==3:
        plt.ylabel(r'$\frac{(g_{k\sigma}(x)-g_{\sigma}(x))}{(k\sigma - \sigma)}$')
    plt.title(r'Difference of Gaussians: $\sigma = 1$, $k = %.1f$' % k[i])
```
